```python
# Import required Libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('files/marketing_AB.csv')

data.head()
```

```python
# check if we have duplicates in user_id 

data.duplicated(subset='user id').sum()
```

```python
# Drop unwanted columns in this case we can drop user_id columns becasue it is unique it doesnot tell anything anymore


data = data.drop(labels='user id', axis=1)
```

```python
# Getting the categroical variables 

data_cat = data[['test group', 'converted', 'most ads day',
       'most ads hour']]
```

```python
# Unique values from the categories

for col in data_cat:
    print(col + ": "  + str(data_cat[col].unique()))
```

### Statistical Tests

```python
from scipy.stats import chi2_contingency

alpha = 0.05

for variable in data_cat.columns:
    if variable != 'converted':
        contingency_table = pd.crosstab(data_cat[variable], data_cat['converted'])
        
        chi2, p, _, _ = chi2_contingency(contingency_table)
        
        print(f"\nChi-squared test for {variable} vs converted: ")
        print(f"Chi-squared value: {chi2}")
        print(f"p-value: {p}")
              
        # check for the significance 
              
        if p < alpha:
            print(f"The difference in conversion rates between {variable} is statistically significant")
        else:
              print(f"There is no significant difference between the conversion rates across {variable}")
```
